Построить график компьютерной реализации модели скользящего среднего первого порядка

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def moving_average(data, window_size):
    cumsum = np.cumsum(data)
    cumsum[window_size:] = cumsum[window_size:] - cumsum[:-window_size]
    return cumsum[window_size - 1:] / window_size
```

```python
np.random.seed(0)
data = np.random.randn(100)
window_size = 10
```

```python
moving_avg = moving_average(data, window_size)
```

```python
plt.figure(figsize=(30, 5))
plt.plot(data, label='Исходные данные')
plt.plot(range(window_size - 1, len(data)), moving_avg, label=f'Скользящее среднее ({window_size}-точечное)')
plt.legend()
plt.grid(True)
plt.show()
```
