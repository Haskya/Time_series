```python
# Построить график компьютерной реализации авторегрессионной модели усовной неоднородности первого порядка и 
# по имеющимся данным оценить параметры модели методом максимального правдоподобия 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model

# Создание искусственных данных (временной ряд)
np.random.seed(0)
n_obs = 100
mean = 0.0  # Среднее значение
volatility = 1.0  # Стандартное отклонение

data = np.random.normal(loc=mean, scale=volatility, size=n_obs)

plt.figure(figsize=(12, 6))
plt.plot(data, label='Искусственные данные')
plt.legend()
plt.show()

model = arch_model(data, vol='Garch', p=1, q=0)
model_fit = model.fit()

print(model_fit.summary())

# график условной дисперсии (график ARCH-эффекта)
model_fit.plot(annualize='D')
plt.show()
```
