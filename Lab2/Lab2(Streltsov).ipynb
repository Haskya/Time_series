```python
# Построить график компьютерной реализации авторегрессионной модели первого порядка 
# и по имеющимся данным оценить параметры модели методом максимального правдоподобия
```

```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Создание искусственных данных (временной ряд)
np.random.seed(0)
n_obs = 100
mean = 0.0  # Среднее значение
volatility = 1.0  # Стандартное отклонение

data = np.random.normal(loc=mean, scale=volatility, size=n_obs)

plt.figure(figsize=(30, 5))
plt.plot(data, label='Исходные данные')
plt.xlabel('Время')
plt.ylabel('Значение')
plt.title('График временного ряда')
plt.legend()
plt.show()


model = ARIMA(data, order=(1, 0, 0))
model_fit = model.fit()
print(model_fit.summary())
```
